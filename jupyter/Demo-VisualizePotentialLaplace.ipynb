```python
from __future__ import division
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

a = 1
V0 = 10
n = 200

V1 = np.zeros(25)

x = np.linspace(-0.5*a, 0.5*a, 25)
y = np.linspace(0, a, 25)

X,Y = np.meshgrid(x,y)

for n in np.arange(1,n,2):
    
    V1 = V1+4*V0/(np.pi*n)*(np.cosh(n*np.pi*X/a))*np.sin(n*np.pi*Y/a)/(np.cosh(n*np.pi/2))
    
#print(V1)
ax.plot_surface(X, Y, V1, rstride=4, cstride=4, color='y')
ax.set_xlabel('x')
ax.set_xlabel('y')
ax.set_zlabel('V')
```
