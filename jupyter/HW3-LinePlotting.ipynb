```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
%matplotlib inline
```

```python
x = np.arange(0,10,0.1)
a = 0.1
b = 0.4
c = 5
y = a*x**2+b*x+c
```

```python
plt.figure(figsize=(8,8))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim([0,10])
plt.ylim([0,20])
```
