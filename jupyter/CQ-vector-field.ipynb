```python
import numpy as np                # imports the numerical library
import matplotlib.pyplot as plt   # imports the plotting library
#plt.style.use('fivethirtyeight')  # sets up pretty plots
%matplotlib inline
x = np.arange(-10,0,2)
y = np.arange(0,6,0.5)
X, Y = np.meshgrid(x,y)
vx = -1*X**2
vy = 0
plt.figure(figsize=(16,8))          # creates the figure object; used when you need a figure (8,8) is just a nice size
plt.quiver(X,Y,vx,vy, color='b', edgecolor='b', units='xy', linewidths=3)              # The call to quiver actually generates the plot
plt.xlim([-10, 0]);                  # Sets the limits of the plot (not needed as Jupyter will autoscale)
plt.ylim([0, 5]);
```
